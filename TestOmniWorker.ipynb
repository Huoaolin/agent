```python
def clu(a, b):
    c = []
    for i in a:
        df = b - i
        if df in a:
            c.append(df)
            return c  # 找到一个就返回
    df = b - max(a)
    print(f"df = {df}")
    if df <= 0:  # 终止条件：差值小于等于 0
        return c
    return clu(a, df)  # 返回递归结果


a = [1, 2, 7]
b = 15
result = clu(a, b)
print(result)
```
